```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
df
#rename column
df= df.rename({'per capita income (US$)':'income'}, axis=1)
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year, df.income,color='blue',marker='*')
```

```python
new_df= df.drop('income', axis='columns')
new_df
```

```python
income= df.income
income
```

```python
# Create linear regression object******
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
m= reg.coef_
m
#m
```

```python
b= reg.intercept_ 
b
#b
```

```python
#y=mx+b
m*2020+b
```
